```python
def findEvenNumbers(digits):
    nums = (max(digits)*111)
    li = []
    for i in range(100,nums):
        st = str(i)
        if int(st[0]) in digits and int(st[1]) in digits and int(st[2]) in digits and int(st)%2 == 0:
            if st.count(st[0]) <= digits.count(int(st[0])) and st.count(st[1]) <= digits.count(int(st[1])) and st.count(st[2]) <= digits.count(int(st[2])):
                li.append(i)
    return li
```
